Here is an exhaustive list of attributes that you can mention while plotting with Matplotlib. These cover all customization options for line plots, markers, and general plot appearance.

1. Data Attributes
x: Data for the x-axis.
y: Data for the y-axis.
2. Line Properties
color or c: Line color.
linewidth or lw: Line thickness.
linestyle or ls: Line style (e.g., '-', '--', ':', '-.').
dash_capstyle: How dashes at the ends of lines appear ('butt', 'round', 'projecting').
dash_joinstyle: How dashes at the joins of lines appear ('miter', 'round', 'bevel').
solid_capstyle: How solid lines ends appear ('butt', 'round', 'projecting').
solid_joinstyle: How solid lines joins appear ('miter', 'round', 'bevel').
antialiased or aa: Whether to smooth lines (True or False).
3. Marker Properties
marker: Marker style (e.g., 'o', 's', '^', '*').
markersize or ms: Marker size.
markeredgecolor: Color of the marker edge.
markeredgewidth: Thickness of the marker edge.
markerfacecolor: Fill color of the marker.
markerfacecoloralt: Alternate fill color for the marker.
markevery: Display markers only at specified intervals (e.g., 1 in 3 or [0, 2, 4]).
4. Color and Style
alpha: Transparency level (0 to 1).
zorder: Order of drawing (higher values are drawn on top).
label: Legend label.
5. General Plot Settings
title: Title of the plot (plt.title()).
xlabel: Label for the x-axis (plt.xlabel()).
ylabel: Label for the y-axis (plt.ylabel()).
grid: Show gridlines (plt.grid()).
xlim: Limit of x-axis (plt.xlim()).
ylim: Limit of y-axis (plt.ylim()).
xticks: Custom ticks for x-axis (plt.xticks()).
yticks: Custom ticks for y-axis (plt.yticks()).
6. Legend Properties
legend: Display a legend (plt.legend()).
loc: Location of the legend (e.g., 'upper left', 'center', 'best').
fontsize: Font size of the legend text.
frameon: Whether to draw a frame around the legend.
shadow: Add a shadow to the legend.
bbox_to_anchor: Control position of the legend box.
7. Error Bars
xerr: Horizontal error bars.
yerr: Vertical error bars.
errorevery: Show errors at every nth data point.
8. Axes and Tick Properties
tick_params: Customize ticks (size, direction, color, etc.).
axhline: Add a horizontal line.
axvline: Add a vertical line.
axhspan: Add a shaded region along the x-axis.
axvspan: Add a shaded region along the y-axis.
9. Advanced Customizations
fill_between: Fill area between two curves.
text: Add text annotations on the plot.
annotate: Add arrows or text to highlight points (xy, xytext).
scatter: Create scatter plots with attributes like c, s, alpha.
10. Other Styling Options
clip_on: Whether the plot should be clipped to the axes bounds.
snap: Align data points to the nearest pixel (True, False).
path_effects: Add effects like shadows, strokes, etc.
picker: Enable picking events for interactive plots.
gid: Set a unique identifier for each artist (useful for custom interactions).

import matplotlib.pyplot as plt

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis')
fig.colorbar(surf)

plt.title("3D Surface Plot")
plt.show()
```

```python
from matplotlib.widgets import Slider

# Interactive plot with slider
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.2)

x = np.linspace(0, 10, 100)
y = np.sin(x)
line, = plt.plot(x, y)

# Slider
ax_slider = plt.axes([0.2, 0.01, 0.65, 0.03])
slider = Slider(ax_slider, 'Frequency', 0.1, 10.0, valinit=1.0)

def update(val):
    freq = slider.val
    line.set_ydata(np.sin(freq * x))
    fig.canvas.draw_idle()

slider.on_changed(update)
plt.show()
```
